```python
import os
sys.path.insert(0, os.path.expanduser('~/python/'))

from warpcmap import warp_colormap, wimshow, wpcolormesh, wpcolor
```

Make some data -- note that the range of the data is (0, 1).

```python
X = rand(20, 20)
```

Display the data twice, with the warped colour map and the original.  Here the warping is chosen to emphasise the values around 0.8. 

```python
z = 0.8
beta = 3
cmap = warp_colormap('jet', z=z, beta=beta)

fig, ax = subplots(1, 2, figsize=(12, 8))

im = ax[0].imshow(X, cmap=cmap)
fig.colorbar(im, ax=ax[0], shrink=0.6)
ax[0].set_title('z = %g  beta = %g' % (z, beta))

im = ax[1].imshow(X, cmap='jet')
fig.colorbar(im, ax=ax[1], shrink=0.6)
ax[1].set_title('No warping')
```

Adjust the range of the data to be between 20 and 30.  Then z corresponds to the proportion of the data range, so that here $z = 0.8$ corresponds to data values around $20 + 0.8 \times (30 - 20) = 28$.

```python
X = X*10 + 20
```

```python
z = 0.8
beta = 3
cmap = warp_colormap('jet', z=z, beta=beta)

fig, ax = subplots(1, 2, figsize=(12, 8))

im = ax[0].imshow(X, cmap=cmap)
fig.colorbar(im, ax=ax[0], shrink=0.6)
ax[0].set_title('z = %g  beta = %g' % (z, beta))

im = ax[1].imshow(X, cmap='jet')
fig.colorbar(im, ax=ax[1], shrink=0.6)
ax[1].set_title('No warping')
```

Use wimshow to set the middle of the colour map directly

```python
vmid = 27
beta = 3

fig, ax = subplots(1, 2, figsize=(12, 8))

im = wimshow(X, ax=ax[0], cmap='plasma', vmid=vmid, beta=beta)
fig.colorbar(im, ax=ax[0], shrink=0.6)
ax[0].set_title('vmid = %g  beta = %g' % (vmid, beta))

im = ax[1].imshow(X, cmap='plasma')
fig.colorbar(im, ax=ax[1], shrink=0.6)
ax[1].set_title('No warping')
```

```python
vmid = 22
beta = 5

fig, ax = subplots(1, 2, figsize=(12, 8))

im = wimshow(X, ax=ax[0], cmap='jet', vmid=vmid, beta=beta)
fig.colorbar(im, ax=ax[0], shrink=0.6)
ax[0].set_title('vmid = %g  beta = %g' % (vmid, beta))

im = ax[1].imshow(X, cmap='jet')
fig.colorbar(im, ax=ax[1], shrink=0.6)
ax[1].set_title('No warping')
```

Use wpcolormesh instead

```python
vmid = 22
beta = 5

fig, ax = subplots(1, 2, figsize=(12, 8))

x = linspace(10, 20, X.shape[1])
y = linspace(-2, 3, X.shape[0])
im = wpcolormesh(x, y, X, ax=ax[0], cmap='jet', vmid=vmid, beta=beta)
fig.colorbar(im, ax=ax[0], shrink=0.6)
ax[0].set_title('vmid = %g  beta = %g' % (vmid, beta))

im = ax[1].pcolormesh(x, y, X, cmap='jet')
fig.colorbar(im, ax=ax[1], shrink=0.6)
ax[1].set_title('No warping')
```

```python
vmid = 28
beta = 3

fig, ax = subplots(1, 2, figsize=(12, 8))

x = linspace(10, 20, X.shape[1])
y = linspace(-2, 3, X.shape[0])
im = wpcolor(x, y, X, ax=ax[0], cmap='cividis', vmid=vmid, beta=beta)
fig.colorbar(im, ax=ax[0], shrink=0.6)
ax[0].set_title('vmid = %g  beta = %g' % (vmid, beta))

im = ax[1].pcolor(x, y, X, cmap='cividis')
fig.colorbar(im, ax=ax[1], shrink=0.6)
ax[1].set_title('No warping')
```
